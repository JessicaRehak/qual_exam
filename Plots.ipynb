```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc,rcParams
```

```python
img_dir = "images/"
label_size = 20
font_size = 18
x_max = 5
plt.figure(figsize=(12,9))
plt.rc('text', usetex=True)
plt.rc('font', family='serif')    
ax = plt.gca()
plt.rc('font', size=font_size)
plt.xticks(fontsize=font_size)
plt.yticks(fontsize=font_size)
plt.xlim([-x_max,x_max])
plt.ylim([-1,2])
plt.xlabel('$n$', fontsize=label_size)
plt.ylabel('$\Lambda(n)/c$', fontsize=label_size)
ax.set_aspect('equal')
#ax.tick_params(axis='y', labelsize=font_size)
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.xticks(np.arange(-x_max,x_max+1, step=1))
plt.grid(True,which='both',color='0.5', linestyle='--')

x = np.linspace(-10,10, 100)
y = np.arctan(x)/x

plt.plot(x,y)

plt.savefig(img_dir + "lambda_function.pdf",format = 'pdf', bbox_inches='tight')
```
